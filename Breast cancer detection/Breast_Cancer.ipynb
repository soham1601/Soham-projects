```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('breast-cancer.csv')
df
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df["diagnosis"]=df["diagnosis"].astype('int64')
```

```python
plt.figure(figsize=(9,3))
df.plot(kind='box', figsize=(20,20))
```

```python
sns.boxplot(df["perimeter_mean"])
```

```python
per25=df["perimeter_mean"].quantile(0.25)
per75=df["perimeter_mean"].quantile(0.75)
iqr=per75-per25
upper=per75+iqr*1.5
lower=per25-iqr*1.5
new_df=df[(df["perimeter_mean"]<upper)&(df["perimeter_mean"]>lower)]
new_df
```

```python
sns.boxplot(df["area_mean"])
```

```python
per25=new_df["area_mean"].quantile(0.25)
per75=new_df["area_mean"].quantile(0.75)
iqr=per75-per25
upper=per75+iqr*1.5
lower=per25-iqr*1.5
new_df1=new_df[(new_df["perimeter_mean"]<upper)&(new_df["perimeter_mean"]>lower)]
new_df1
```

```python
per25=new_df1["perimeter_worst"].quantile(0.25)
per75=new_df1["perimeter_worst"].quantile(0.75)
iqr=per75-per25
upper=per75+iqr*1.5
lower=per25-iqr*1.5
new_df2=new_df1[(new_df1["perimeter_worst"]<upper)&(new_df1["perimeter_worst"]>lower)]
new_df2
```

```python
per25=new_df2["area_worst"].quantile(0.25)
per75=new_df2["area_worst"].quantile(0.75)
iqr=per75-per25
upper=per75+iqr*1.5
lower=per25-iqr*1.5
new_df3=new_df2[(new_df2["area_worst"]<upper)&(new_df2["area_worst"]>lower)]
new_df3
```

```python
fig, ax = plt.subplots(figsize=(20,20))
sns.heatmap(new_df3.corr(),cmap="YlGnBu", annot=True, ax=ax)
```

```python
new_df3
```

```python
new_df3.columns
```

```python
final_df=new_df3[["radius_mean","perimeter_mean","area_mean","concavity_mean","concave points_mean","radius_worst","perimeter_worst","area_worst","concave points_worst","diagnosis"]].copy()
```

```python
final_df
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python

scaler=MinMaxScaler()
scaler.fit(final_df)
scaled=scaler.transform(final_df)
df_min_max=pd.DataFrame(data=scaled,columns=["radius_mean","perimeter_mean","area_mean","concavity_mean","concave points_mean","radius_worst","perimeter_worst","area_worst","concave points_worst","diagnosis"])
```

```python
df_min_max
```

```python
from sklearn.model_selection import train_test_split
x = df_min_max.drop(['diagnosis'], axis=1)
y = df_min_max.diagnosis
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
regressor=LogisticRegression()
regressor.fit(xtrain,ytrain);
y_pred=regressor.predict(xtest)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(ytest,y_pred)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
naive=GaussianNB()
naive.fit(xtrain,ytrain)
y_pred=naive.predict(xtest)
```

```python
metrics.accuracy_score(ytest,y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=5,metric="minkowski", p=1)
knn.fit(xtrain,ytrain)
y_pred=knn.predict(xtest)
```

```python
metrics.accuracy_score(y_pred,ytest)
```
